```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open('calamardo.jpg')
```

```python
img_array = np.array(img)
print(np.shape(img_array)) 
```

# **RANDOM IMAGE (NOISE)**

```python
img_ran = np.random.randint(0,150,(255,255,3))  # (MIN, MAX, SHAPE)
```

```python
plt.imshow(img_ran)
plt.show()
```

# **MESHGRID (DEGRADADO)**

```python
x = np.linspace(0, 255, 255)    # (MIN, MAX, LENGHT)
y = np.linspace(0, 255, 255)
C, R = np.meshgrid(x, y)           
```

```python
plt.imshow(R, cmap = 'gray')   # cmap is necessary
plt.show()
```

```python
plt.imshow(C, cmap = 'gray')
plt.show()
```

```python
loops_x = 3    # loops in x axis            also try with negative values :v
loops_y = 5    # same with y axis

sin = np.cos((loops_x*2*np.pi/255)*R + (loops_y*2*np.pi/255)*C)
```

```python
plt.imshow(sin, cmap = 'gray')
plt.show()
```

# **Position of Image**

```python
img_array[100, 150, 0]   # [X, Y, CHANNEL]
```

```python
im_R = img_array.copy()
im_R[:, :, (1, 2)] = 0      # CHANNELS G-B NULLS
im_G = img_array.copy()
im_G[:, :, (0, 2)] = 0      # CHANNELS R-B NULLS
im_B = img_array.copy()
im_B[:, :, (0, 1)] = 0      # CHANNELS R-G NULLS

im_RGB = np.concatenate((im_R, im_G, im_B), axis = 1)
```

```python
plt.imshow(im_RGB)
plt.show()
```

```python
im_rgb = []
for i in [(1,2),(0,2),(0,1)]:
  im = img_array.copy()
  im[:, :, i] = 0
  im_rgb.append(im)

im_rgb = np.concatenate((im_rgb), axis = 1)
```

```python
plt.imshow(im_rgb)
plt.show()
```
