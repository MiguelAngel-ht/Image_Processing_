# **Laplaciano**

Esta operacin reconoce los bordes en una imagen, ya que la derivada de una funcin es el cambio de la pendiente en cada punto, y la segunda derivada tiene valores nulos en los puntos crticos. Los cambios de intensidades de la imagen como se ve en un histograma se pueden detectar con la segunda derivada. El laplaciano es la extensin del concepto de segunda derivada com ms dimensiones y matemticamente se expresa como:

$\nabla ^2 f(x,y) = \frac{\partial^2 f(x,y)}{\partial x^2} + \frac{\partial^2 f(x,y)}{\partial y^2} $

```python
img_lap = cv.Laplacian(img,cv.CV_64F)
```

```python
plt.imshow(img_lap,'gray')
plt.show()
```
